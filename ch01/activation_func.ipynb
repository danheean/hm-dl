```python
# 필요한 라이브러리들을 임포트
import numpy as np  # 수치 계산을 위한 numpy
import matplotlib.pyplot as plt  # 그래프 그리기를 위한 matplotlib
from scipy.special import expit  # 시그모이드 함수를 위한 scipy

# x축 데이터 생성: -10부터 10까지 0.2 간격으로 배열 생성
x = np.arange(-10, 10, 0.2)
print(x)  # 생성된 x 배열 출력
```

```python
# 1x2 서브플롯 생성 (가로 12, 세로 4 크기)
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# 왼쪽 그래프: 시그모이드 함수 그리기
axs[0].plot(x, expit(x))
axs[0].set_title('sigmoid function')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')

# 오른쪽 그래프: ReLU 함수 그리기 (x.clip(0)은 x가 0보다 작으면 0으로, 크면 x값 그대로)
axs[1].plot(x, x.clip(0))
axs[1].set_title('relu function')
axs[1].set_xlabel('x')
axs[1].set_ylabel('f(x)')

# 그래프 표시
plt.show()
```
